```python
# Buscando un caso similar del ITEM faltante para saber si rellenarlo o dropearlo

Filtered_dimProd1 = dimProduct[dimProduct["ITEM_DESCRIPTION"] == "CLORALEX EL RENDIDOR BOT PLAST 2LT"]
Filtered_dimProd1
```

```python

Filtered_dimProd2 = dimProduct[dimProduct["ITEM_DESCRIPTION"] == "CLOROX MASCOTAS BLANQUEADOR+DETERGENTE GALON 10L"]
Filtered_dimProd2
```

```python
# No se encontraron coincidencias, se dejara el valor con "NO DEFINIDO"

dimProduct["ITEM"] = dimProduct["ITEM"].fillna("NO DEFINIDO")

# Cambiando el valor null de ATTR1 y ATTR3 a No definido como se muestra en algunos otros casos

dimProduct["ATTR1"] = dimProduct["ATTR1"].fillna("NO DEFINIDO")
dimProduct["ATTR3"] = dimProduct["ATTR3"].fillna("NO DEFINIDO")
```

```python
# Renaming columns to better Merge DF
dimCategory.rename(columns={"CATEGORY" : "CATEGORY_NAME"}, inplace = True)
dimCategory.rename(columns={'ID_CATEGORY' : "CATEGORY"}, inplace = True)
dimCategory.head()
```

```python
productCategory = pd.merge(dimProduct, dimCategory, on='CATEGORY', how='inner')
productCategory
```

```python
productCatSeg = pd.merge(productCategory, dimSegment[["CATEGORY", "FORMAT", "SEGMENT"]], on='CATEGORY', how='inner')
productCatSeg
```

```python
productCatSeg.rename(columns={"FORMAT_x" : "FORMAT1", "FORMAT_y" : "FORMAT2"}, inplace = True)
productCatSeg
```

```python
# Reorder Columns
productCatSeg = productCatSeg[["MANUFACTURER", "BRAND", "ITEM", "ITEM_DESCRIPTION", "CATEGORY", "CATEGORY_NAME", "FORMAT1", "FORMAT2", "ATTR1", "ATTR2", "ATTR3", "SEGMENT"]]
productCatSeg.head()
```

```python
calendarSales = pd.merge(dimCalendar, factSales,  on='WEEK', how='inner')
calendarSales.rename(columns={"ITEM_CODE" : "ITEM"}, inplace = True)
calendarSales.head()
```

```python
productSales = pd.merge(calendarSales, productCatSeg , on="ITEM", how="inner")
productSales.head()
```

```python
productCatSeg.to_csv('productCatSeg.csv', index=False)
calendarSales.to_csv('calendarSales.csv', index=False)
productSales.to_csv('productSales.csv', index=False)
```
